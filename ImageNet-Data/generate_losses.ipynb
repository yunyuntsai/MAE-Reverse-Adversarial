## KS Test

```python
from scipy.stats import ks_2samp
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open("output/base_loss.pkl", "rb") as f:
    base_loss_values = pickle.load(f)

with open("output/fgsm_4.pkl", "rb") as f:
    fgsm4_loss_values = pickle.load(f)

with open("output/fgsm_8.pkl", "rb") as f:
    fgsm8_loss_values = pickle.load(f)

with open("output/fgsm_16.pkl", "rb") as f:
    fgsm16_loss_values = pickle.load(f)
```

```python
ks_2samp(base_loss_values[0:16], fgsm4_loss_values[0:16], alternative="greater")
# if p-value greater than some threshold

# M2, known clean vs unknown adversarial
# mix the samples for clean and base 
# load directories separately when testing and concat them together randomly to shuffle
#           torch concat dataset function? torch.utils.data.concatdataset
#           shuffled once, not shuffled once (argument)
```

```python
ks_2samp(base_loss_values[0:16], fgsm8_loss_values[0:16], alternative="greater")
# if p-value greater than some threshold
```

```python
ks_2samp(base_loss_values[0:16], fgsm16_loss_values[0:16], alternative="greater")
# if p-value greater than some threshold
```

```python
np.linspace(0, 1, len(base_loss_values))
```

```python
def ks_plot_comp(data_1, data_2):
    plt.figure(figsize=(12, 7))
    plt.scatter(data_1, range(len(data_1)), label="base loss")
    plt.scatter(data_2, range(len(data_2)), label="attack loss")
    plt.legend('top right')
    plt.title('Comparing 2 data distributions for KS-Test')
    plt.show()
```

```python
ks_plot_comp(base_loss_values, fgsm4_loss_values)

# label base value as 0, adversarial as 1, calculate p-value, 
# predict labels in the middle will be different
# if p-value of batched sample then we predict label as 1 or 0
```

```python
ks_plot_comp(base_loss_values, fgsm8_loss_values)
```

```python
ks_plot_comp(base_loss_values, fgsm16_loss_values)
```
